```python
import numpy as np
import pandas as pd
```

```python
crypto = pd.read_csv('combined_crypto')
crypto.drop(['Unnamed: 0'], axis=1, inplace=True)
crypto.head()
```

```python
twitter = pd.read_csv('combined_twitter')
twitter.drop(['Unnamed: 0'], axis=1, inplace=True)
twitter.head()
```

```python
combined = pd.merge(crypto, twitter, how='left', on = ['date', 'symbol'])
combined
```

```python
combined.to_csv('combined_final')
```
